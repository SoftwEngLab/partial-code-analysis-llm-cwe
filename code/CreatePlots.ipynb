```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_accuracy_from_csv(csv_file):

    df = pd.read_csv(csv_file)
    
    required_columns = {"Experiment", "Accuracy (%)", "Pairwise Accuracy (%)"}
    if not required_columns.issubset(df.columns):
        raise ValueError(f"CSV file must contain columns: {required_columns}")
    
    experiments = df["Experiment"].tolist()
    accuracy = df["Accuracy (%)"].tolist()
    pairwise_accuracy = df["Pairwise Accuracy (%)"].tolist()
    

    x = np.arange(len(experiments))
    width = 0.35


    fig, ax = plt.subplots(figsize=(10, 6))


    ax.bar(x - width/2, accuracy, width, label='Accuracy (%)')
    ax.bar(x + width/2, pairwise_accuracy, width, label='Pairwise Accuracy (%)')


    ax.set_xlabel('Prompting Strategy')
    ax.set_ylabel('Accuracy (%)')
    ax.set_title('Accuracy and Pairwise Accuracy by Experiment')
    ax.set_xticks(x)
    ax.set_xticklabels(experiments, rotation=45, ha='right')
    ax.legend()


    plt.tight_layout()
    plt.show()

```

```python
plot_accuracy_from_csv("prelim_results/Experiment_Results_cwe_078_bedrock.csv")
```
